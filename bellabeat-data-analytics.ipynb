### ▣ **4.4 Defining the data bias and fairness :**

Data bias is a type of error that systematically skews the results in a certain direction. It occurs when data points at a certain preference in a favour of audience or against audience, a person or thing. It must be showing the equal amount of the people no matter what. It doesn't need to reinforce the partiality. It can be categorised into three types: **Observer Bias**, **Interpretation Bias**, and **Confirmation Bias**. 

Once the data represent the whole population by all means, it brings the **fairness** to the analysis. However, for this data **(Fitbit data)**, due to the limitation of size (30 users) and not having any demographic information we could encounter a sampling bias. We are not sure if the sample is representative of the population as a whole. Another problem we would encounter is that the dataset is not current and also the time limitation of the survey (2 months long). That is why we will give our case study an operational approach.

### ▣ **4.5 Defining the clean and bad data source through ROCCC :**

A data source can be very complex and frustrating when it is not organized well. There are many data sources that are not reliable, current, cited and original. To identify the good data source, we need to follow a certain process given below and based on our Fitbit Fitness Tracker Data, we can define whether data is collected from clean or bad source; 
* **ROCCC for Fitbit Fitness Data:**
   + **R** : Realiable | biased data, not sure if the data is vetted
   + **O** : Original | yes, can locate the original public data
   + **C** : Comprehensive | no, many informations are missing such as age, weight, height, gender, etc also missing the completeness and accuracy. 
   + **C** : Current | no, this is a historical data source (04/12/2016 - 05/12/2016)
   + **C** : Cited | yes

✔ -Here, we need to determine the type of users spend more time in the bed before falling a sleep and their sleep records with the data we have. Healthy sleep is an essential part of life. Trying to establish a normal sleep pattern is critical to everyday functioning. Normal sleep for adults means that you fall asleep within 10 to 20 minutes and get about 7–8 hours a night.  

We will classify the users by their sleeping behaviour considering the following parameters;

* **Users sleep too soon** - Users sleep so fast after getting in the bed or (just within a couple of minutes after getting in the bed)
* **Users sleep on time** - Users sleep on time (in between 10-20 minutes of recommended ideal time)
* **Users take extra time** - Users take more time than recommended before falling asleep ( after 10-20 minutes of recommended ideal time)

For a detailed information, [click here](https://www.healthline.com/health/healthy-sleep/how-long-does-it-take-to-fall-asleep#takeaway)
* **

```r
head(new_sleep_day)
```

```r
sleep_behaviour <- new_sleep_day %>%
    group_by(id) %>%
    mutate(time_diff = total_time_in_bed - total_minutes_asleep) %>%
    summarize(avg_time_diff = mean(time_diff)) %>%
    mutate(sb_class = case_when(
          avg_time_diff < 10 ~ "Users sleep too soon",
          avg_time_diff >= 10 & avg_time_diff <=20 ~ "Users sleep on time",
          avg_time_diff > 20 ~ "Users take extra time"
          ))                                                                         #sb=sleeping behaviour                     
head(sleep_behaviour)
```

After classifying the users, we need to create their percentage value for a better understanding.

```r
sb_percent <- sleep_behaviour %>%
    group_by(sb_class) %>%
    summarize(sb_total = n()) %>%
    mutate(sb_overall = sum(sb_total))%>%
    group_by(sb_class) %>%
    summarize(sb_class_value = sb_total/sb_overall) %>%
    mutate(sb_class_percent = scales :: percent(sb_class_value))

head(sb_percent)
```

We notice that sb_class_percent values are a character data type we can either leave it as it is and ordered them manually for your plot. Here we are going to manually represent them by executing the code. Later on, we will use another technique in the `geom_text()` to convert the value in percentage. 

```r
sb_percent$sb_class = factor(sb_percent$sb_class, levels = c("Users sleep on time", "Users sleep too soon", "Users take extra time"))

sb_piechart <- sb_percent %>%
    ggplot(aes(x = "", y = sb_class_value, fill = sb_class)) +
    geom_col(color = "white") +
    coord_polar("y", start = 0) +
    geom_text(aes(label = sb_class_percent), position = position_stack(vjust = 0.5), size = 6, fontface = "bold", color = "white") +
    scale_fill_manual(values = c("#facbc1","#f6b3a4", "#f68d77")) +
    labs(title = "Classification of users taking time to fall asleep", 
       subtitle = "Ideal time of 10-20 minutes to be taken for anyone in the bed to fall asleep", 
       caption = paste0("Fitbit Fitness Tracker Data from: ", min_date, " to ", max_date)) + 
    theme(panel.background = element_blank(),
        axis.line = element_blank(),
        axis.text = element_blank(),  
        axis.ticks = element_blank(),
        axis.title = element_blank(),
        plot.title = element_text(hjust = 0.1, size = 19, face = "bold", color = "#f68d77"),
        plot.subtitle = element_text(hjust = 0.3, face = "bold", size = 13),
        plot.caption = element_text(vjust = 3.5, size = 12, color = "#f68d77", face = "bold"))

print(sb_piechart)
```

* **5000-7499 steps a day (low active)**
* **7500-9999 steps a day (somewhat active)**
* **10,000-12,499 steps a day (active)**
* **>12,500 steps.d (highly active)**

Classification has been made per the following article <https://pubmed.ncbi.nlm.nih.gov/14715035/>

First we will calculate the daily steps average by user.

```r
steps_class_df <- daily_activity %>%   # Categorizing the users type based on their average steps/day and recommended steps/day
    group_by(id) %>%
    summarize(avg_act = mean(total_steps)) %>%
    mutate(avg_act_class = case_when(
        avg_act < 5000 ~ "sedentary",
        avg_act >= 5000 & avg_act <= 7499 ~ "low active",
        avg_act >= 7500 & avg_act <= 9999 ~ "somewhat active",
        avg_act >= 10000 & avg_act <= 12499 ~ "active",
        avg_act > 12500 ~ "very active" 
        ))

head(steps_class_df)

steps_class_percent <- steps_class_df %>% # determining the weightage of the each class/type a user belongs to.
    group_by(avg_act_class) %>%
    summarize(total_act = n()) %>%
    mutate(overall_total = sum(total_act)) %>%
    group_by(avg_act_class) %>%
    summarize(activity_step_percent = total_act/overall_total)

head(steps_class_percent)

steps_class_percent$avg_act_class <- factor(steps_class_percent$avg_act_class, levels = c("very active","active","somewhat active","low active","sedentary"))


piechart_avg_steps <- steps_class_percent %>%
    ggplot(aes(x = "", y = activity_step_percent, fill = avg_act_class)) +
    geom_col(color = "#f68d77") + 
    coord_polar("y", start = 0) + 
    geom_text(aes(label = scales:: percent(activity_step_percent)), position = position_stack(vjust = 0.5), fontface = "bold", color = "white", size = 5) +
    scale_fill_grey(name = "Type of Users", breaks = c("very active", "active", "somewhat active", "low active", "sedentary"), 
                    labels = c("very active: >12500", "active: 12499-10000", "somewhat active: 9999-7500", "low active: 7499-5000", "sedentary: <5000")) +
    labs(title = "Classification of users average steps per day", 
       subtitle = "On the average, users steps/day divided into five categories", 
       caption = paste0("Fitbit Fitness Tracker Data from: ", min_date, " to ", max_date)) + 
    theme(panel.background = element_blank(),
        axis.line = element_blank(),
        axis.text = element_blank(),  
        axis.ticks = element_blank(),
        axis.title = element_blank(),
        plot.title = element_text(size = 19, face = "bold", color = "#f68d77"),
        plot.subtitle = element_text(face = "bold", size = 13, hjust = 0.40),
        plot.caption = element_text(hjust = 2,size = 12, color = "#f68d77", face = "bold"),
        legend.title = element_text(size = 13, face = "bold"),
        legend.text = element_text(size = 11))

print(piechart_avg_steps)
```

#### **<span style="color:#f68d77">⑥ Users's Intensities level:</span>**
* **
Physical activity is defined as any voluntary bodily movement produced by skeletal muscles that requires energy expenditure. Physical activity encompasses all activities, at any intensity, performed during any time of day or night. It includes both exercise and incidental activity integrated into daily routine. To understand these user's activity level, we will need to set a parameter and we will classify these users based on the **<span style = "color:#f68d77">METs (Metabolic Equivalents for tasks)</span>**. One MET is defined as the energy you use when you're resting or sitting still. An activity that has a value of 4 METs means you're exerting four times the energy than you would if you were sitting still. Since this dataset is not comphrehensive, we will need to define users activities level or intensities based on recommended METs. [click here for more details.](https://www.cdc.gov/nccdphp/dnpa/physical/pdf/pa_intensity_table_2_1.pdf) 

✔ Almost every occupation requires some mix of light, moderate, or vigorous activities, depending on the task at hand.

* **Light Intensity Level:** Users with METs lower than 3.0 (<3.0)
* **Moderate Intensity Level:** Users with METs in between 3.0-6.0
* **Vigorous Intensity Level:** Users with METs higher than 6.0 (>6.0)

To follow along with question: how much METs a user need during their activity level, we are going to use `daily_activity` and merge two datesets; `hourly_calories` & `hourly_intensities`

```r
head(daily_activity)

piechart_mets_weightage <- daily_activity %>%
    mutate(total_min = very_active_minutes + fairly_active_minutes + lightly_active_minutes) %>%
    group_by(id) %>%
    summarize(avg_int = mean(total_min), avg_cal = mean(calories)) %>%
    mutate(mets = avg_cal/avg_int) %>%
    mutate(mets_class = case_when(
        mets < 3 ~ "Light Intensity, METs: < 3.0",
        mets >= 3 & mets <= 6 ~ "Moderate Intensity, METs: 3.0-6.0",
        mets > 6 ~ "Vigorous Intensity, METs: >6.0")) %>%
    group_by(mets_class) %>%
    summarize(total = n()) %>%
    mutate(overall_total = sum(total)) %>%
    group_by(mets_class) %>%
    summarize(percent = total/overall_total) %>% 
    mutate(mets_class = factor(mets_class, levels = c("Light Intensity, METs: < 3.0", "Moderate Intensity, METs: 3.0-6.0", "Vigorous Intensity, METs: >6.0"))) %>%
ggplot(aes(x = "", y = percent, fill = mets_class)) + 
geom_col(color = "white") + 
coord_polar("y", start = 0) +
geom_text(aes(label = scales :: percent(percent)), position = position_stack(vjust = 0.5), fontface = "bold", size = 6, color = "white") + 
scale_fill_manual(values = c("#80cbfa", "#f68d77", "#5c5a59"), name = "Percent Color Scale") +
    labs(title = "Classification of users on their average METs", subtitle = "On the average, users METs divided in three categories:Light,Moderate and Vigorous", 
       caption = paste0("Fitbit Fitness Tracker Data from: ", min_date, " to ", max_date)) + 
    theme(panel.background = element_blank(),
        axis.line = element_blank(),
        axis.text = element_blank(),  
        axis.ticks = element_blank(),
        axis.title = element_blank(),
        plot.title = element_text(size = 19, face = "bold", color = "#f68d77", hjust = 6.5),
        plot.subtitle = element_text(size = 13, face = "bold"),
        plot.caption = element_text(hjust = 2,size = 12, color = "#f68d77", face = "bold"),
        legend.title = element_text(size = 13, face = "bold"),
        legend.text = element_text(size = 12))

print(piechart_mets_weightage)
```

**✒Insight 16:** Based on the calculation with given datasets, these users's average METs based on the recommended METs are clearly visible now. #### **<span style="color:#f68d77">⑦ Users smart devices usage based on days AND hours used for smart devices:</span>**
* **
Here we will figure out how many days and how many hours a user use his smart devices. To accomplish this, we will further classify these users through setting a parameter. 

* **Low Usage:** Users who use smart devices throughout a range of 1-10 days a month
* **Medium Usage:** Users who use smart devices throughout a range of 11-20 days a month
* **High Usage:** Users who use smart devices throughout a range of 21-31 days a month

Since it is a long data, we will count the id. 

```r
head(daily_activity)

monthly_days_usage_df <- daily_activity %>% # Creating separate df to use it later for filtering the category
    group_by(id) %>%
    summarize(days_usage = sum(n())) %>%
    mutate(usage_cat = case_when(
        days_usage >= 1 & days_usage <= 10 ~ "Low Usage: 1-10 days",
        days_usage >= 11 & days_usage <=20 ~ "Medium Usage: 11-20 days",
        days_usage >= 21 & days_usage <= 31 ~ "High Usage: 21-31 days"))
    
calcu_monthly_days <- monthly_days_usage_df %>% 
group_by(usage_cat) %>%
    summarize(total = n()) %>%
    mutate(overall_total = sum(total)) %>%
    group_by(usage_cat) %>%
    summarize(usage_percent = total/overall_total) %>%
    mutate(usage_class = factor(usage_cat, levels = c("High Usage: 21-31 days", "Medium Usage: 11-20 days", "Low Usage: 1-10 days")))

bargraph_monthlydays <- calcu_monthly_days %>%
ggplot(aes(x = usage_class, y = usage_percent, fill = usage_percent)) + 
geom_col() +
geom_text(aes(label = scales :: percent(usage_percent)), fontface = "bold", color = "#5c5a59", vjust = 1.3, size = 6) + 
labs(title = "Number of days using smart devices", subtitle = "Days used classification based on the parameter: low, medium & high", x = "Usage Classification", y = "Usage Percent", 
     caption = paste0("Fitbit Fitness Tracker Data: ", min_date, " to ", max_date)) +
scale_fill_gradient(low = "#80cbfa", high = "#f68d77") +
theme(plot.title = element_text(size=19, color = "#f68d77", face = "bold"), 
    plot.subtitle = element_text(size=13, color = "black", face = "bold"), 
    plot.caption = element_text(size=12, color = "#f68d77", face = "bold"),
    axis.text.x = element_text(size=12, color = "#5c5a59"),
    axis.text.y = element_text(size=14, color = "#5c5a59"),
    axis.title.x = element_text(size=14, color = "#5c5a59", face = "bold", vjust = 0.2), 
    axis.title.y = element_text(size=14, color = "#5c5a59", face = "bold"),
    strip.text.x = element_text(size=13), 
    strip.text.y = element_text(size=13),
    legend.title = element_text(size=12), 
    legend.text = element_text(size=10)) +
easy_remove_legend() +
scale_y_continuous(label = percent)

print(bargraph_monthlydays)
```
